```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])       
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels, title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('V Kohli vs RG Sharma')
plt.xlabel('season')
plt.ylabel('Runs scored')
```

```python
# colors(hex) and line(width and style) and marker(size)

#plt.plot(batsman['index'],batsman['V Kohli'],color='green')
#plt.plot(batsman['index'],batsman['RG Sharma'],color='black')

plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F') #using hex code
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6')

plt.title('V Kohli vs RG Sharma')
plt.xlabel('season')
plt.ylabel('Runs scored')
```

```python
#linestyle
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='dashed') #default linestyle is solid
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dotted')
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#linewidth
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#marker,markersize
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)  #D=diamond (D,+,o,.,>,<,....)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

#plt.legend() bydefault, loc='best' means automatically figure out best place 
plt.legend(loc='upper right')
```

```python
# size parameter
tips = sns.load_dataset('tips')


# slower
#plt.scatter(tips['total_bill'],tips['tip'])
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20) #3D info

#Matplotlib makes it simple to change the plot size for all points in a scatter plot.
#The size of points is based on the s= parameter.
```

```python
# scatterplot using plt.plot

plt.plot(tips['total_bill'],tips['tip'])  #2D line plot
```

```python
plt.plot(tips['total_bill'],tips['tip'],'o') #gives scatter plot
# plt.plot vs plt.scatter
# but no other parameters in scatter plot available using this method
#but it is faster than scatter plot.. so prefered for large dataset
```

## Bar chart

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='Green') #vertical barchart
```

```python
# horizontal bar chart        if there r more no. of categories
plt.barh(colors,children,color='blue')
```

```python
# color and label
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
#multiple plot
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])

#overlapping
#no direct fn
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
plt.bar(df['batsman'],df['2015'],width=1)
```

```python
plt.bar(df['batsman'],df['2015'],width=.5)
```

```python
# Multiple Bar charts
# xticks

plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# a problem...... if category names are big
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
```

```python
plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
```

```python
# Stacked Bar chart..................bottom parameter
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

## Pie Chart

```python
# simple data
data = [23,45,100,20,49]

plt.pie(data)
plt.show()
```

```python
# labeling
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'])
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.4,0,0,0,0,0],shadow=True)
plt.show()
```
